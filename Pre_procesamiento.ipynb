```python
import pandas as pd
import kaggle
import numpy as np
kaggle.api.dataset_download_files('sagarikashreevastava/cognitive-distortion-detetction-dataset', path='data', unzip=True)
```

```python
d1 = pd.read_csv('data/Annotated_data.csv')
d1['Texto'] = d1['Patient Question'] + [x if isinstance(x, str) else ' ' for x in d1['Distorted part']]
d1['Dominant Distortion'] = d1['Dominant Distortion'].replace(distorsiones_cognitivas_traducidas)
d1['Secondary Distortion (Optional)'] = d1['Secondary Distortion (Optional)'].replace(distorsiones_cognitivas_traducidas)
d1['Distorsión'] = d1.apply(lambda x: ','.join([str(y) for y in [x['Dominant Distortion'], x['Secondary Distortion (Optional)']] if pd.notna(y)]), axis=1)
d1['Dataset'] = 'D1'

def separar_y_expandir(df, columna):
    df[columna] = df[columna].str.split(',')
    df_expandido = df.explode(columna).reset_index(drop=True)

    return df_expandido
d1 = separar_y_expandir(d1, 'Distorsión')
```

```python
df_final = d_comp;
```

```python
labels = df_final['Distorsión'].str.split(',', expand=True).stack()
frecuencia = labels.value_counts().reset_index()
frecuencia.columns = ['Distorsión', 'Frecuencia']
frecuencia["Frecuencia (%)"] = frecuencia["Frecuencia"]/frecuencia["Frecuencia"].sum()*100
frecuencia
```

```python
etiquetas_baja_frecuencia = frecuencia[frecuencia['Frecuencia (%)'] < 3]['Distorsión']
def filtro_etiquetas(row):
  etiquetas = row['Distorsión'].split(',')
  return not any(etiqueta in etiquetas_baja_frecuencia.values for etiqueta in etiquetas)

df_filtrado = df_final[df_final.apply(filtro_etiquetas, axis=1)]
df_filtrado = df_filtrado.sort_values(by='Dataset')
df_filtrado.reset_index(drop=True, inplace=True)

labels = df_filtrado['Distorsión'].str.split(',', expand=True).stack()
frecuencia = labels.value_counts().reset_index()
frecuencia.columns = ['Distorsión', 'Frecuencia']
frecuencia["Frecuencia (%)"] = frecuencia["Frecuencia"]/frecuencia["Frecuencia"].sum()*100
frecuencia
```

```python
def eliminar_filas(df, num_valores_finales, string_eliminar):
    filas_con_string = df[df['Distorsión'].str.contains(string_eliminar, na=False)]
    filas_a_eliminar = filas_con_string.sample(frac=1).index.tolist()
    num_filas_a_eliminar = len(filas_a_eliminar) - num_valores_finales
    filas_a_eliminar = filas_a_eliminar[:num_filas_a_eliminar]
    df = df.drop(filas_a_eliminar)
    return df

prueba = df_filtrado

for cad in frecuencia["Distorsión"]:
  prueba = eliminar_filas(prueba, 231, cad)
```

```python
labels = prueba['Distorsión'].str.split(',', expand=True).stack()
frecuencia = labels.value_counts().reset_index()
frecuencia.columns = ['Distorsión', 'Frecuencia']
frecuencia["Frecuencia (%)"] = frecuencia["Frecuencia"]/frecuencia["Frecuencia"].sum()*100
frecuencia
```
